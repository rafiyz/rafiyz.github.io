### Python Logical Operators
Logical operators are used to combine conditional statements:

- "and" Returns True if both statements are true "x < 5 and x < 10"
- "or" Returns True if one of the statements is true "x < 5 or x < 4"
- "not" Reverse the result, returns False if the result "not(x < 5 and x < 1 is true)


```python
x=4
y=19

# example of "and"
if x<3 and x<y:
    print (True)
else:
    print (False)



# example of "or"
if x<3 or x<y:
    print (True)
else:
    print (False)


# example of "not"
if not(x<5 and x<y):
    print (True)
else:
    print (False)



# example of "not" with different input
if not(x<2 and x>y):
    print (True)
else:
    print (False)
```

### Python Bitwise Operators
Bitwise operators are used to compare (binary) numbers:

- "&" (AND) Sets each bit to 1 if both bits are 1 x & y
- "|" (OR) Sets each bit to 1 if one of two bits is 1 x | y
- "^" (XOR) Sets each bit to 1 for different bits and 0 for the same bits x^y
- "~" (NOT) Sets the opposite of the value 

```python
# Bitwise AND example
a = 12  #    Binary: 1100
b = 5   #    Binary: 0101
# Output: 4 (Binary: 0100)

result = a & b
print(f"{a} & {b} = {result}") 
```

```python
# Bitwise OR example
a = 12  #    Binary: 1100
b = 5   #    Binary: 0101
#Output: 13 (Binary: 1101)
result = a | b
print(f"{a} | {b} = {result}")  
```

```python
# Bitwise XOR example
a = 12   # Binary: 1100
b = 5    # Binary: 0101
#Output 9 (Binary: 1001)

result = a ^ b
print(f"{a} ^ {b} = {result}")  
```

```python
# Bitwise NOT example
a = 12  # Binary: 1100
# Output: -13 (Binary: Two's complement of 1100)

result = ~a
print(f"~{a} = {result}")  
```

### Operator Precedence
Operator precedence describes the order in which operations are performed.

```python
#Example
#Parentheses has the highest precedence, meaning that expressions inside parentheses must be evaluated first:

print((6 + 3) - (6 + 3))
```
