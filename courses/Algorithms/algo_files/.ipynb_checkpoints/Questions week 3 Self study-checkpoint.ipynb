### **Python Assignment: Operators**

# **1. Python Membership Operators (`in`, `not in`)**

## **Q1: Check if a word exists in a sentence**

sentence = "Python is a powerful programming language."
- Check if the word "Python" is present using a membership operator.
- Print a message indicating whether the word exists or not.

## **Q2: Check if a number exists in a list**

numbers = [10, 20, 30, 40, 50]
- Take an integer input from the user.
- Check if the input number is in the list using a membership operator.
- Print the result.

# **2. Python Assignment Operators (`=`, `+=`, `-=`, `*=`, `/=`, `//=`, `%=`, `**=`)**

## **Q3: Demonstrate assignment operators**

x = 10
- Use different assignment operators (`+=`, `-=`, `*=`, `/=`, `//=`, `%=`, `**=`) to modify `x`.
- Print the updated value after each operation.

## **Q4: Modify one number using another**

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
- Modify `a` using `b` with assignment operators.
- Print the result after each modification.

# **3. Python Bitwise Operators (`&`, `|`, `^`, `~`, `<<`, `>>`)**

## **Q5: Perform bitwise operations**

a = 5  # 0b0101
b = 3  # 0b0011
- Apply bitwise AND, OR, XOR, and NOT operations.
- Print the results for each operation.

## **Q6: Bitwise shift operations**

num = int(input("Enter a number: "))
- Perform left shift (`<< 2`) and right shift (`>> 2`) operations on `num`.
- Print the results.

# **4. Operator Precedence**

## **Q7: Evaluate expressions manually and with Python**

- Evaluate the following expressions manually, then verify using Python:
  ```python
  result1 = 10 + 5 * 2
  result2 = (10 + 5) * 2
  result3 = 10 + 5 * 2 ** 3
  result4 = 10 / 2 * 5 + 3 - 1
  ```
- Compare the manual calculations with Python output.

## **Q8: Demonstrate operator precedence with user input**

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))
- Apply a combination of arithmetic operations (`+, -, *, /, **`).
- Observe the effect of operator precedence on the result.
