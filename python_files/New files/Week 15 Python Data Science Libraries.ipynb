```python
# Importing NumPy

import numpy as np
```

```python
# Creating NumPy arrays
arr = np.array([1, 2, 3, 4, 5])
print("NumPy Array:", arr)
```

```python
# Array indexing and slicing
print("First element:", arr[0])
print("Sliced array:", arr[1:4])
```

```python
# Array operations (addition, multiplication, broadcasting)
arr2 = np.array([10, 20, 30, 40, 50])
print("Addition:", arr + arr2)
print("Multiplication:", arr * 2)
```

```python
# Statistical functions
print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr))
```

```python
# Linear algebra with NumPy
matrix = np.array([[1, 2], [3, 4]])
print("Matrix:", matrix)
print("Matrix Transpose:", np.transpose(matrix))
print("Matrix Inverse:", np.linalg.inv(matrix))
```

```python
pip install pandas
```

```python
# Importing pandas
import pandas as pd
```

```python
# Series and DataFrame creation
series = pd.Series([10, 20, 30, 40])
df = pd.DataFrame({"Name": ["Alice", "Bob"], "Age": [25, 30]})
print("Series:\n", series)
print("DataFrame:\n", df)
```

```python
# Reading/writing data (CSV)
df.to_csv('example.csv', index=False)  # Save DataFrame to CSV without row indices
new_df = pd.read_csv('example.csv')    # Read CSV back into a DataFrame

print("New DataFrame read from CSV:\n", new_df)
```

```python
# Indexing, selecting, and filtering data
print("Selecting column:", df['Name'])
print("Filtering data:\n", df[df['Age'] > 26])
```

```python
#Advance Level
# Handling missing data
data = {"A": [1, 2, np.nan], "B": [5, np.nan, np.nan]}
df_missing = pd.DataFrame(data)
print("Data with missing values:\n", df_missing)
print("Fill missing values:\n", df_missing.fillna(0))
```

```python
#Advance Level
# Data aggregation and grouping
group_data = {"Category": ['A', 'B', 'A', 'B'], "Values": [10, 20, 30, 40]}
group_df = pd.DataFrame(group_data)
print("Grouped Sum:\n", group_df.groupby('Category').sum())
```

```python
#Advance Level
# Merging and joining datasets
df1 = pd.DataFrame({'ID': [1, 2], 'Name': ['Alice', 'Bob']})
df2 = pd.DataFrame({'ID': [1, 2], 'Score': [85, 90]})
merged_df = pd.merge(df1, df2, on='ID')
print("Merged DataFrame:\n", merged_df)
```

```python
#Advance Level
# Pivot tables
pivot_data = {"Name": ['Alice', 'Bob', 'Alice'], "Subject": ['Math', 'Math', 'English'], "Score": [90, 80, 95]}
pivot_df = pd.DataFrame(pivot_data)
print("Pivot Table:\n", pivot_df.pivot_table(index='Name', columns='Subject', values='Score'))
```

```python
# Plotting with Matplotlib
plt.plot([1,2,3,4], [1,4,9,16])
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
# Customizing plots
plt.plot([1,2,3,4], [1,4,9,16], color='red', linestyle='--', marker='o')
plt.title("Customized Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(["Squares"])
plt.show()
```

```python
# Loading real-world dataset (Titanic dataset)
titanic = sns.load_dataset('titanic')
print(titanic.head())
```

```python
# Data cleaning and preprocessing
titanic['age'] = titanic['age'].fillna(titanic['age'].median())
```

```python
# Exploratory Data Analysis (EDA)
print(titanic.describe())
sns.countplot(x='sex', data=titanic)
plt.title("Passenger Count by Gender")
plt.show()
```

```python
# Visual storytelling
sns.barplot(x='class', y='fare', data=titanic)
plt.title("Average Fare by Class")
plt.show()
```

```python
# Simple predictive modeling (optional)
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# Selecting features and target
features = titanic[['age', 'fare']].fillna(0)
target = titanic['survived']
```

```python
# Splitting data
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
# Training model
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Model accuracy
print("Model Accuracy:", model.score(X_test, y_test))
```
